```python
import pandas as pd
from pickle import load
from IPython.display import display
```

```python
publi = load( open('data-clean/dic-Publicidad2QuienEsQuien.p', 'rb') )
publi = set(publi.values())
type(publi), len(publi)
```

```python
sexenios = ['FOX','CAL','EPN']
colors_sex = {'FOX':'red','CAL':'green','EPN':'blue'}
colors_met = {'direct':'#e7298a','limited':'#a6761d','open':'#8da0cb'}
```

## ¿Cuantas empresas de publicidad estan en las favoritas?

```python
for key in sexenios:
    print(key)
    df = pd.read_csv('data-clean/{}-favoritas.csv'.format(key))
    sel = publi.intersection( set(df['empresa']) )
    sel = df[df['empresa'].isin(sel)].sort_values('monto_total', ascending=False)
    m,t = sel.shape[0], df.shape[0]
    print('\tFavoritas {} de {} ({}%)'.format(m,t,round(m*100/t,3)))
    display(sel)
```

## ¿Cuantas empresas de publicidad estan en las puede con todas?

```python
for key in sexenios:
    print(key)
    df = pd.read_csv('data-clean/{}-puedecontodas.csv'.format(key))
    sel = publi.intersection( set(df['empresa']) )
    sel = df[df['empresa'].isin(sel)].sort_values('monto_total', ascending=False)
    m,t = sel.shape[0], df.shape[0]
    print('\tPuede con todas {} de {} ({}%)'.format(m,t,round(m*100/t,3)))
    display(sel.head(25))
```
