```python
import pandas as pd
from pickle import load
from IPython.display import display
```

```python
publi = load( open('data-clean/dic-Publicidad2QuienEsQuien.p', 'rb') )
publi = set(publi.values())
type(publi), len(publi)
```

```python
sexenios = ['FOX','CAL','EPN']
colors_sex = {'FOX':'red','CAL':'green','EPN':'blue'}
colors_met = {'direct':'#e7298a','limited':'#a6761d','open':'#8da0cb'}
```

## ¿Cuantas empresas de publicidad estan en las favoritas?

```python
for key in sexenios:
    print(key)
    df = pd.read_csv('data-clean/{}-favoritas.csv'.format(key))
    sel = publi.intersection( set(df['empresa']) )
    sel = df[df['empresa'].isin(sel)].sort_values('monto_total', ascending=False)
    m,t = sel.shape[0], df.shape[0]
    print('\tFavoritas {} de {} ({}%)'.format(m,t,round(m*100/t,3)))
    display(sel)
```

## ¿Cuantas empresas de publicidad estan en las puede con todas?
Estas empresas trabajan con contratos de hasta 300 con 85 instituciones en la administración de Fox, lo cual es un signo de concentración de presupuesto en dichas empresas a través de varias instituciones.Este patrón no parece estar tan acentuado para las administraciones de Calderón y EPN, sin embargo tambien se observan dinámicas de concentración de contratos en proveedores como la UNAM con 216 contratos con 50 instituciones en la admin de Calderón, y Milenio diario en la admin de EPN.

Cabe destacar, que aunque el comportamiento de las favoritas en el ramo de publicidad se mantuvo igual en los tres sexenios, durante el sexenio de EPN el número de empresas "puede con todas" en materia de publicidad se duplico. Dada la gran cantidad de contratos y montos asignados a estas empresas, esto muestra que las empresas de publicidad fueron favorecidas de manera destacable durante el sexenio de EPN a nivel federal.

```python
for key in sexenios:
    print(key)
    df = pd.read_csv('data-clean/{}-puedecontodas.csv'.format(key))
    sel = publi.intersection( set(df['empresa']) )
    sel = df[df['empresa'].isin(sel)].sort_values('monto_total', ascending=False)
    m,t = sel.shape[0], df.shape[0]
    print('\tPuede con todas {} de {} ({}%)'.format(m,t,round(m*100/t,3)))
    display(sel.head(25))
```
